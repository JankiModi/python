```python
import matplotlib.pyplot as mp
import numpy as np
ypoints = np.array([3,8,1,7])
mp.subplot(1,2,1)
mp.plot(ypoints,'p-.k',ms='18',mec='k',mfc='hotpink',lw='3')
# (marker shape,style,color), marker size = '18', marker edge color='red', marker face color = 'blue', linewidth='3'

font1 = {'family':'serif','color':'darkred','size':20}
mp.title('stars',fontdict = font1, loc='left')

mp.xlabel('eee')
mp.ylabel('ooo')

mp.grid() #add grid lines 
mp.grid(axis='y') # x or y or both(default)

mp.grid(color='k',linestyle='--',linewidth='1')
# grid(color = 'color', linestyle = 'linestyle', linewidth = number).

font2 = {'family':'serif','color':'blue'}
mp.subplot(1,2,2)
mp.plot(np.array([3,5,2,7]))
mp.title('blabla',loc = 'left')

mp.suptitle('super title',fontdict=font2)

mp.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])
sizes = np.array([20,50,100,200,500,100,60,90,10,300,600,800,75])

plt.scatter(x, y, c=colors, cmap='viridis',s=sizes,alpha = 0.5)
#alpha for transperancy 
plt.colorbar()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))
plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')
plt.colorbar()
plt.show()
```
